```python
import json
import pandas as pd
from io import StringIO

with open("results_19.json", "r") as f:
    data = json.load(f)

# extract the log text
log_text = data["activitiesLog"]

# turn the semicolon-separated text into a dataframe
df = pd.read_csv(StringIO(log_text), sep=";")

df
```

```python
eme_results = df[df["product"] == "EMERALDS"].copy()
tom_results = df[df["product"] == "TOMATOES"].copy()
```
