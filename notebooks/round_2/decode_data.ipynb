```python
import pandas as pd

day_0_prices = pd.read_csv("../../data/round_2/prices_round_2_day_0.csv", sep=";")
day_1_prices = pd.read_csv("../../data/round_2/prices_round_2_day_1.csv", sep=";")
day_neg_1_prices = pd.read_csv("../../data/round_2/prices_round_2_day_-1.csv", sep=";")

day_0_trades = pd.read_csv("../../data/round_2/trades_round_2_day_0.csv", sep=";")
day_1_trades = pd.read_csv("../../data/round_2/trades_round_2_day_1.csv", sep=";")
day_neg_1_trades = pd.read_csv("../../data/round_2/trades_round_2_day_-1.csv", sep=";")

day_0_ash_prices = day_0_prices[day_0_prices["product"] == "ASH_COATED_OSMIUM"].copy()
day_1_ash_prices = day_1_prices[day_1_prices["product"] == "ASH_COATED_OSMIUM"].copy()
day_neg_1_ash_prices = day_neg_1_prices[day_neg_1_prices["product"] == "ASH_COATED_OSMIUM"].copy()

day_0_root_prices = day_0_prices[day_0_prices["product"] == "INTARIAN_PEPPER_ROOT"].copy()
day_1_root_prices = day_1_prices[day_1_prices["product"] == "INTARIAN_PEPPER_ROOT"].copy()
day_neg_1_root_prices = day_neg_1_prices[day_neg_1_prices["product"] == "INTARIAN_PEPPER_ROOT"].copy()

ash_prices = pd.concat([day_neg_1_ash_prices, day_0_ash_prices, day_1_ash_prices])
root_prices = pd.concat([day_neg_1_root_prices, day_0_root_prices, day_1_root_prices])

ash_prices["global_time"] = ash_prices["day"] * 1_000_000 + ash_prices["timestamp"]
root_prices["global_time"] = root_prices["day"] * 1_000_000 + root_prices["timestamp"]
```
