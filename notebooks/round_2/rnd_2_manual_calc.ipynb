```python
import numpy as np

def research(x):
    research_pnl = 200_000 * np.log(1 + x) / np.log(1 + 100)
    print(research_pnl)

def scale(x):
    scale = .07 * x
    print(scale)
```

```python
research(18)
scale(82)
```

```python
from scipy.optimize import minimize_scalar
import numpy as np

def f(x):
    return -(200_000 * np.log(1 + x) / np.log(101) + 0.07 * x)  # negative for maximization

res = minimize_scalar(f, bounds=(0, 100), method='bounded')

print("Optimal x:", res.x)
print("Max value:", -res.fun)
```
