```python
import json

def extract_sandbox_logs(input_file, output_file):
    with open(input_file, "r", encoding="utf-8") as f:
        lines = f.readlines()

    json_objects = []
    current_block = []
    inside_sandbox = False

    for line in lines:
        stripped = line.strip()

        if stripped.startswith("Sandbox logs"):
            inside_sandbox = True
            continue

        if stripped.startswith("Activities log") or stripped.startswith("Trade History"):
            inside_sandbox = False

        if not inside_sandbox:
            continue

        if stripped:
            current_block.append(stripped)

            if stripped == "}":
                try:
                    obj = json.loads("\n".join(current_block))

                    # 🔥 CRITICAL FIX
                    obj["sandboxLog"] = obj.get("sandboxLog") or " "
                    obj["lambdaLog"] = obj.get("lambdaLog") or " "

                    json_objects.append(obj)

                except json.JSONDecodeError as e:
                    print("Skipping bad block:", e)

                current_block = []

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(json_objects, f, indent=2)

    print(f"✅ Extracted {len(json_objects)} logs → {output_file}")


extract_sandbox_logs(
    "2026-04-05_11-02-27.log",
    "sandbox_logs.json"
)
```
